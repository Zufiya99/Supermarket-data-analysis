## Sales Representation and Market Share Analysis

```python
sale.groupby(["Location"]).count()
```

```python
sale.groupby(["Location","Gender"]).count()
```

```python
sale.groupby(["Location","Gender"]).count()["Invoice ID"]
```

```python
location_sale = sale.groupby(["Location","Gender"]).count()["Invoice ID"]
```

```python
location_sale
```

```python
# unstack method --> (Used for multiple indexes)
```

```python
unstack_sale = location_sale.unstack(level=0)
```

```python
unstack_sale
```

```python
unstack_sale.plot(kind="bar")
```

```python
members = sale.groupby(["Member","Location"]).count()["Invoice ID"]
```

```python
members
```

```python
members.unstack(level=0).plot(kind="bar")
plt.show()
```
